```python
# python code to extract the youtube video title and link from the provided playlist url and return response in markdown form 
import yt_dlp
def extract_youtube_playlist_info(playlist_url):
    ydl_opts = {
        'extract_flat': True,
        'quiet': True,
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        playlist_info = ydl.extract_info(playlist_url, download=False)

    markdown_response = ""
    for entry in playlist_info['entries']:
        title = entry.get('title', 'No title available')
        video_url = entry.get('url', 'No URL available')
        markdown_response += f"- [{title}]({video_url})\n"

    return markdown_response

# Example usage
if __name__ == "__main__":
    #playlist_url = ""
    #playlist_url = "https://youtube.com/playlist?list=PLyqSpQzTE6M9LibNqhhCYLDvfVGhcRDN4&feature=shared"
    playlist_url = "https://youtube.com/playlist?list=PLHZhjPByiV3K3PCGrCiTLbgq8o-Mbhn3b&feature=shared"
    response = extract_youtube_playlist_info(playlist_url)
    print(response)
```
